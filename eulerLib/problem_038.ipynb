# Project Euler: Problem 038: Pandigital Multiples
<a href="https://projecteuler.net/problem=38">Task definition</a>

"Take the number $192$ and multiply it by each of $1, 2,$ and $3$:

$$192√ó1	=192$$
$$192√ó2	=384$$
$$192√ó3	=576$$
 
By concatenating each product we get the $1$ to $9$ pandigital, $192384576$. We will call $192384576$ the concatenated product of $192$ and ($1,2,3$).

The same can be achieved by starting with $9$ and multiplying by $1, 2, 3, 4,$ and $5,$ giving the pandigital, $918273645$, which is the concatenated product of $9$ and ($1,2,3,4,5$).

What is the largest $1$ to $9$ pandigital $9$-digit number that can be formed as the concatenated product of an integer with ($1,2,\dots,n$) where $n > 1$?"

```python
def is_pandigital_1to9(s):
    return len(s) == 9 and set(s) == set("123456789")
```

##### Where's the limit

$1 \times 4$-digtis number will still be $4$-digits.
$2 \times 4$-digtis number can reach a $5$-digits number. Plus those first $4$-digits, we would have a $9$-digits number.

Therefore our limit will be $10000$.

```python
limit = 10000

nbr = 2
maxStr = "0"
maxNbr = 0
maxCount = 0

while nbr < limit:
    txt = ""
    count = 0

    for i in range(1, 10):
        entry = nbr * i
        txt += str(entry)

        count += 1

        if len(txt) > 9:
            break
        
        if len(txt) == 9 and is_pandigital_1to9(txt):
            if int(txt) > int(maxStr):
                maxStr = txt
                maxNbr = nbr
                maxCount = count

    nbr += 1

print(f"{maxStr = }\n{maxNbr = }\n{maxCount = }")
```
