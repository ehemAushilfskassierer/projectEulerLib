# Project Euler: Problem 025: 1000-digit Fibonacci Number
<a href="https://projecteuler.net/problem=25">Task definition</a>

"The Fibonacci sequence is defined by the recurrence relation:

$F_n = F_{n-1} + F_{n-2}$, where $F_1 = 1$ and $F_2 = 1$.
Hence the first $12$ terms will be:

$$F_1 = 1$$
$$F_2 = 1$$
$$F_3 = 2$$
$$F_4 = 3$$
$$F_5 = 5$$
$$F_6 = 8$$
$$F_7 = 13$$
$$F_8 = 21$$
$$F_9 = 34$$
$$F_{10} = 55$$
$$F_{11} = 89$$
$$F_{12} = 144$$
 
The $12$ th **term**, $F_{12}$, is the first term to contain three digits.

What is the index of the first term in the Fibonacci sequence to contain $1000$ digits?"

```python
lGoal = 1000
```

```python
def calcIndexOfFibonacci(l: int = 10,
                         bShowNbr: bool = False) -> int:
    """
    Calculate the the Fibonacci number with "l" digits.
    Count the amount of calculations to get there (index).

    Parameters:
        l (int): Number of digits the number should have.
        bShowNbr (bool): Print a text with the number
                         and index.
    Returns:
        int: Index of number with "l" digits 
    """
    if not isinstance(l, int):
        raise TypeError("l must be an integer")
    if not l >= 1:
        raise ValueError("l must be >= 1")
    if not isinstance(bShowNbr, bool):
        raise TypeError("bShowNbr must be a bool")
    
    a, b = 1, 1

    count = 1

    if l == 1:
        return count
    
    while len(str(a)) != l:
        a, b = b, a + b
        count += 1

    if bShowNbr:
        print(f"The first Fibonacci with {l} digits is "
              f"{a} and has the index {count}.")
        
    return count
```

```python
calcIndexOfFibonacci(3, True)
```

```python
calcIndexOfFibonacci(lGoal)
```
