# Step-by-Step

2. Create a list of UK coin denominations in descending order: 200, 100, 50, 20, 10, 5, 2, 1.

3. Create a dictionary mapping each coin value to its human-readable name (e.g., 200 â†’ "2Â£", 50 â†’ "50p", etc.).

4. Initialize an empty list to store all valid combinations that will be found.

5. Define an inner recursive function whose task is to build coin combinations:

   * If the remaining amount becomes 0, it means a valid combination has been found.
   * The valid combination is copied and stored in the results list.
   * If enabled, the combination is printed in readable form.

6. Inside the recursive function, loop over the coin list starting from the provided index:

   * Skip coins that are larger than the remaining amount.
   * Increase the count of the selected coin in the current combination.
   * Recursively call the function with the decreased remaining amount and the same index (because coins can be reused).
   * After returning, decrease the count again to undo the choice (backtracking).

7. Create the initial data structure that holds each coin paired with a count starting at zero.

9. Start the recursive backtracking process using the full target amount, starting at the first coin.

10. After all recursive paths finish, return the number of valid combinations that were collected.

```python
def calcCoinSums(target: int, bShowComb: bool = False) -> int:
    """
    Calculate the number of unique combinations of UK coins that sum to a given target amount.

    This function computes all possible combinations of standard UK coin denominations
    (2Â£, 1Â£, 50p, 20p, 10p, 5p, 2p, 1p) that add up to the provided target amount.
    The target may be given as an integer or float representing pounds (Â£). Internally,
    the target is converted into pence to operate using integer arithmetic.

    Parameters
    ----------
    target : int or float
        The monetary target in pounds (Â£). Must be a positive number.
        Examples: 1, 2.50, 0.35

    bShowComb : bool, optional
        If True, print every valid coin combination in the format "count x coin".
        Default is False.

    Returns
    -------
    int
        The number of distinct combinations of UK coins that sum exactly to the target.

    Raises
    ------
    TypeError
        If `target` is not an int/float, or if `bShowComb` is not a boolean.

    ValueError
        If `target` is less than or equal to zero.

    Notes
    -----
    The algorithm uses a recursive backtracking approach to explore all valid
    coin combinations. Coins are processed in non-increasing order to ensure
    that combinations are counted without duplicates (i.e., order does not matter).

    Examples
    --------
    >>> calcCoinSums(1)
    4563

    >>> calcCoinSums(0.05, bShowComb=True)
    3
    """
    if not isinstance(target, (int, float)):
        raise TypeError("target must be an float (or int)")
    if target <= 0:
        raise ValueError("target must be > 0")
    if not isinstance(bShowComb, bool):
        raise TypeError("bShowComb must be a bool")
    
    target_pennies = int(target*100)
    coins = [200, 100, 50, 20, 10, 5, 2, 1]
    names = {200:"2Â£",100:"1Â£",50:"50p",20:"20p",10:"10p",5:"5p",2:"2p",1:"1p"}
    
    results = []

    def backtrack(remaining, start, combo):
        if remaining == 0:
            results.append(combo.copy())
            if bShowComb:
                print(" + ".join(f"{cnt}x{names[c]}" for c, cnt in combo if cnt > 0))
            return
        
        for i in range(start, len(coins)):
            coin = coins[i]
            if coin > remaining:
                continue
            combo[i][1] += 1
            backtrack(remaining - coin, i, combo)
            combo[i][1] -= 1

    combo = [[coin, 0] for coin in coins]
    backtrack(target_pennies, 0, combo)

    return len(results)
        
```
