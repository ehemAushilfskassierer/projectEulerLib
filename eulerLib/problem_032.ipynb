# Project Euler: Problem 032: Pandigital Products
<a href="https://projecteuler.net/problem=32">Task definition</a>

"We shall say that an $n$-digit number is pandigital if it makes use of all the digits $1$ to $n$ exactly once; for example, the $5$-digit number, $15234$, is $1$ through $5$ pandigital.

The product $7254$ is unusual, as the identity, $39 \times 186 = 7254$, containing multiplicand, multiplier, and product is $1$ through $9$ pandigital.

Find the sum of all products whose multiplicand/multiplier/product identity can be written as a $1$ through $9$ pandigital.

*HINT: Some products can be obtained in more than one way so be sure to only include it once in your sum*."

```python
def pandigitalProductSum(bShowComb: bool = False) -> int:
    """
    Compute the sum of all unique products whose multiplicand/multiplier/product
    identity can be written as a 1-9 pandigital number.

    A 1-9 pandigital identity uses each digit from 1 through 9 exactly once across
    the concatenation of multiplicand, multiplier, and product. For example:

        39 x 186 = 7254   ‚Üí   "391867254" is 1-9 pandigital.

    This function identifies all such valid identities and returns the sum of all
    *distinct* products. (Some products can be formed in multiple ways but are only
    counted once.)

    Parameters
    ----------
    bShowComb : bool, optional
        If True, print every valid pandigital identity in the format "a x b = product".
        Default is False.

    Returns
    -------
    int
        The sum of all unique products appearing in pandigital identities.

    Raises
    ------
    TypeError
        If `bShowComb` is not a boolean.

    Notes
    -----
    The search ranges follow from digit-length constraints:
    - multiplicand can be 1 or 2 digits.
    - multiplier can be 3 or 4 digits.
    - product is always 4 digits.
    The total concatenation must use exactly nine digits.

    Examples
    --------
    >>> pandigitalProductSum()
    45228

    >>> pandigitalProductSum(bShowComb=True)
    39 x 186 = 7254
    4 x 1738 = 6952
    ...
    """
    if not isinstance(bShowComb, bool):
        raise TypeError("bShowComb must be a bool")
```

#### Call function with Parameter bShowComb `True`

```python
pandigitalProductSum(True)
```
