# Project Euler: Problem 033: Digit Cancelling Fractions
<a href="https://projecteuler.net/problem=33">Task definition</a>

"The fraction $49/98$ is a curious fraction, as an inexperienced mathematician in attempting to simplify it may incorrectly believe that $49/98 =4/8$, which is correct, is obtained by cancelling the $9$ s.

We shall consider fractions like, $30/50 =3/5$, to be trivial examples.

There are exactly four non-trivial examples of this type of fraction, less than one in value, and containing two digits in the numerator and denominator.

If the product of these four fractions is given in its lowest common terms, find the value of the denominator."

```python
def curious_fractions_denominator(show_details: bool = False) -> int:
    """
    Compute the denominator of the product of all non-trivial 'curious fractions'.

    A curious fraction is a two-digit fraction numerator/denominator where
    cancelling a shared digit yields an equivalent simplified fraction, even
    though such cancellation is mathematically invalid. For example:

        49 / 98 â†’ cancelling the 9's gives 4 / 8, and indeed 49/98 = 4/8.

    This function finds all such non-trivial digit-cancelling fractions,
    multiplies them together, reduces the resulting product to its simplest
    terms, and returns the *denominator* of that reduced product.

    Parameters
    ----------
    show_details : bool, optional
        If True, print each curious fraction found and its cancellation step.
        Default is False.

    Returns
    -------
    int
        The denominator of the product of all curious fractions
        (in lowest terms).

    Raises
    ------
    TypeError
        If `show_details` is not a boolean.

    Notes
    -----
    The search space includes all two-digit numerators and denominators.
    Trivial examples with zeros (e.g., 30/50) are excluded.

    Examples
    --------
    >>> curious_fractions_denominator()
    100

    >>> curious_fractions_denominator(show_details=True)
    Found: 49/98 = 4/8
    Found: 19/95 = 1/5
    ...
    """
    if not isinstance(show_details, bool):
        raise TypeError("show_details must be a bool")
```
