# Project Euler: Problem 036: Double-base Palindromes
<a href="https://projecteuler.net/problem=36">Task definition</a>

"The decimal number, $585 =1001001001_2$ (binary), is palindromic in both bases.

Find the sum of all numbers, less than one million, which are palindromic in base $10$ and base $2$.

(Please note that the palindromic number, in either base, may not include leading zeros.)"

```python
for i in range(10):
    print(f"{i = } = {bin(i)}")
```

```python
def check_palindrom(n: int) -> bool:
    return str(n) == str(n)[::-1]
```

```python
check_palindrom(9)
```

```python
check_palindrom(bin(9)[2:])
```

```python
dctRes = {}
lstRes = []

for i in range(1000000):
    if check_palindrom(i) and check_palindrom(bin(i)[2:]):
        dctRes[i] = bin(i)[2:]
        lstRes.append(i)

dctRes
```

```python
sum(lstRes)
```
