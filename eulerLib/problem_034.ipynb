# Project Euler: Problem 034: Digit Factorials
<a href="https://projecteuler.net/problem=34">Task definition</a>

"$145$ is a curious number, as $1!+4!+5! =1 +24 +120 =145$.

Find the sum of all numbers which are equal to the sum of the factorial of their digits.

Note: As $1! =1$ and $2! =2$ are not sums they are not included."

```python
import math

dct_factorials = {str(i): math.factorial(i)\
                  for i in range(10)}

dct_factorials
```

```python
def is_factorion(n):
    return n == sum(dct_factorials[d] for d in str(n))
```

```python
upperLimit = sum(dct_factorials[str(i)]\
                 for i in range(10)) + 1

lowerLimit = max(10, sum(dct_factorials[str(i)]\
                 for i in range(3)))
```

```python
res = [n for n in range(lowerLimit, upperLimit)\
       if is_factorion(n)]

res
```

```python
sum(res)
```
