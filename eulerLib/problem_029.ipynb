# Project Euler: Problem 029: Distinct Powers
<a href="https://projecteuler.net/problem=29">Task definition</a>

"Consider all integer combinations of $a^b$ for $2 \le a \le 5$ and $2 \le b \le 5$:

<table>
  <tr>
    <td>2^2 = 4</td>
    <td>2^3 = 8</td>
    <td>2^4 = 16</td>
    <td>2^5 = 32</td>
  </tr>
  <tr>
    <td>3^2 = 9</td>
    <td>3^3 = 27</td>
    <td>3^4 = 81</td>
    <td>3^5 = 243</td>
  </tr>
  <tr>
    <td>4^2 = 16</td>
    <td>4^3 = 64</td>
    <td>4^4 = 256</td>
    <td>4^5 = 1024</td>
  </tr>
  <tr>
    <td>5^2 = 25</td>
    <td>5^3 = 125</td>
    <td>5^4 = 625</td>
    <td>5^5 = 3125</td>
  </tr>
</table>

If they are then placed in numerical order, with any repeats removed, we get the following sequence of $15$ distinct terms:

$$
4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125.
$$

How many distinct terms are in the sequence generated by $a^b$ for $2 \le a \le 100$ and $2 \le b \le 100$?"

```python
def countDistinctTerms(aLimit, bLimit, bShowTerms=False) -> int:
    """
    Count the number of distinct terms of a^b for the ranges:
        aLimit = (a_min, a_max)
        bLimit = (b_min, b_max)

    Parameters
    ----------
    aLimit : tuple(int, int)
        A tuple (start, end) defining the inclusive range for 'a'.
        Must be two integers where start < end.

    bLimit : tuple(int, int)
        A tuple (start, end) defining the inclusive range for 'b'.
        Must be two integers where start < end.

    bShowTerms : bool, optional
        If True, print the set of all distinct terms generated.

    Returns
    -------
    int
        The number of unique values of a**b over the given ranges.

    Raises
    ------
    TypeError
        If inputs are not tuples of integers.
    ValueError
        If the ranges are invalid (start >= end).
    """

    # ---- Input validation ----
    if (
        not isinstance(aLimit, tuple) or
        not isinstance(bLimit, tuple) or
        len(aLimit) != 2 or
        len(bLimit) != 2
    ):
        raise TypeError("aLimit and bLimit "
                        "must be tuples of two integers.")

    if not all(isinstance(x, int) for x in aLimit + bLimit):
        raise TypeError("All values in aLimit and bLimit "
                        "must be integers.")

    if aLimit[0] >= aLimit[1]:
        raise ValueError(f"aLimit start ({aLimit[0]}) "
                         f"must be < end ({aLimit[1]}).")

    if bLimit[0] >= bLimit[1]:
        raise ValueError(f"bLimit start ({bLimit[0]}) "
                         f"must be < end ({bLimit[1]}).")

    # ---- Compute distinct terms ----
    terms = {a ** b for a in range(aLimit[0], aLimit[1] + 1)
             for b in range(bLimit[0], bLimit[1] + 1)}
    
    if bShowTerms:
        print(terms)

    return len(terms)
```

#### Test the function

```python
bLim = (2, 5)
aLim = (2, 5)

countDistinctTerms(aLim, bLim, True)
```

#### Solve the task

```python
aLim = (2, 100)
bLim = aLim

countDistinctTerms(aLim, bLim)
```
