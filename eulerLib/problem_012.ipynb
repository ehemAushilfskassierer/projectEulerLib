# Project Euler: Problem 012: Highly Divisible Triangular Number
<a href="https://projecteuler.net/problem=12">Task definition</a>


"The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 
1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:
| Triangle Number| Divisors     |
| -------------- | ------------- |
| **1** | 1 |
| **3** | 1, 3 |
| **6** | 1, 2, 3, 6 |
| **10** | 1, 2, 5, 10 |
| **15** | 1, 3, 5, 15 |
| **21** | 1, 3, 7, 21 |
| **28** | 1, 2, 4, 7, 14, 28 |

We can see that 28 is the first triangle number to have over five divisors.
What is the value of the **first triangle number to have over five hundred divisors**?"

```python
def count_divisors(n):
    divisors = 0
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            divisors += 2 if i != n // i else 1
    return divisors
```

```python
def triangle_number(n):
    return n * (n + 1) // 2
```

```python
n = 1
while True:
    tri = triangle_number(n)
    if count_divisors(tri) > 500:
        print(f"The first triangular number with more than 500 divisors is: {tri}")
        break
    n += 1
```
