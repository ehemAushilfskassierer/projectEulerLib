# Project Euler: Problem 023: Non-Abundant Sums
<a href="https://projecteuler.net/problem=23">Task definition</a>

"A perfect number is a number for which the sum of its proper divisors is exactly equal to the number. For example, the sum of the proper divisors of $28$  would be $1 + 2 + 4 + 7 + 14 = 28$, which means that $28$ is a perfect number.

A number $n$ is called deficient if the sum of its proper divisors is less than $n$ and it is called abundant if this sum exceeds $n$.

As $12$ is the smallest abundant number, $1 + 2 + 3 + 4 + 6 = 16$, the smallest number that can be written as the sum of two abundant numbers is $24$. By mathematical analysis, it can be shown that all integers greater than $28123$ can be written as the sum of two abundant numbers. However, this upper limit cannot be reduced any further by analysis even though it is known that the greatest number that cannot be expressed as the sum of two abundant numbers is less than this limit.

Find the sum of all the positive integers which cannot be written as the sum of two abundant numbers."

|Expression |Definition|Example|
| :---: | :------ | :------ |
|**Deficient number**|*The sum of proper divisors of $n$ is* **less** *than $n$.*|$4$ is **deficient**, since $1 + 2 = 3$.|
|**Perfect number**|*The sum of proper divisors of $n$ is* **equal** *to $n$.*|$28$ is **perfect**, since $1 + 2 + 4 + 7 + 14 = 28$.|
|**Abundant number**|*The sum of proper divisors of $n$ is* **greater** *than $n$.*|$12$ is **abundant**, since $1 + 2 + 3 + 4 + 6 = 16$.|

##### Create function to define the type of number

```python
def defineNbr(n: int) -> int:
    """
    Define if a number is:
        deficient (0), perfect (1), abundant (2), or not defined (3).

    Parameters:
        n (int): A positive integer (>=1) to check.

    Returns:
        int: Classification code:
            0 --> deficient
            1 --> perfect
            2 --> abundant
            3 --> not defined
    """
    if not isinstance(n, int):
        raise TypeError("Input n must be an integer")
    if n < 1:
        raise ValueError("Input n must be >= 1")

    # Compute proper divisors
    divisors = [1]  # 1 is always a proper divisor
    for i in range(2, n // 2 + 1):
        if n % i == 0:
            divisors.append(i)

    sum_div = sum(divisors)

    # Classification
    if sum_div < n:
        return 0  # deficient
    elif sum_div == n:
        return 1  # perfect
    elif sum_div > n:
        return 2  # abundant
    else:
        return 3  # not defined (shouldn't happen)
```

##### Create a function to loop through every number til limit

```python
def calcSumOfNonAbundant(limit: int = 28123) -> int:
    """
    Calculate the sum of all numbers up to the limit
    that cannot be expressed as the sum of
    two abundant numbers.

    Parameters:
        limit (int): Integer for upper limit
                     to check and sum.

    Returns:
        int: The total sum of numbers not expressible
             as the sum of two abundant numbers
             up to the limit.
    """
    if not isinstance(limit, int):
        raise TypeError("limit must be an integer")
    if not limit > 1:
        raise ValueError("limit must be >= 1")
    
    if limit > 28123:
        print(f"Limit {limit} set to 28123. All larger"
              " numbers can be expressed as the sum"
              " of two abundant numbers.")
        limit = 28123

    # Get all abundant numbers up to the limit
    abundants = [i for i in range(1, limit + 1)
                 if defineNbr(i) == 2]
    
    print(f"{len(abundants) = }")
    print(f"{max(abundants) = }")
    print(f"{min(abundants) = }")

    can_be_written = [False] * (limit + 1)

    for i in range(len(abundants)):
        for j in range(i, len(abundants)):
            s = abundants[i] + abundants[j]
            if s <= limit:
                can_be_written[s] = True
            else:
                break
    print("Amount of numbers that can be expressed as a sum: " +
          f"{sum(can_be_written)}")
    
    noSumOfTwoAbundatns = [i for i in range(1, limit + 1)
                           if not can_be_written[i]]
    
    print(f"{noSumOfTwoAbundatns = }")

    total = sum(noSumOfTwoAbundatns)
    print(f"Sum of non-abundant sums up to {limit}:")

    return total
```

```python
calcSumOfNonAbundant()
```
